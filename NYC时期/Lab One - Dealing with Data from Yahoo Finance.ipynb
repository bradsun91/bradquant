```python
import pandas as pd
import pandas_datareader.data as wb
from datetime import datetime
import matplotlib.pyplot as plt

%matplotlib inline
```

### Getting raw data (Adj Close) from Yahoo Finance

```python
msft = wb.DataReader('MSFT', data_source = 'yahoo', start = datetime(2016,1,1), end = datetime(2016,8,19))
data = pd.DataFrame({'msft':msft['Adj Close']})
msft_adj_close = msft['Adj Close']
m = msft_adj_close
m.plot(figsize = (16,8))
```

### Calculate the percentage change on a daily basis

```python
dt_pct_chg = data.pct_change()
dt_pct_chg.plot(figsize = (16,8))
```
