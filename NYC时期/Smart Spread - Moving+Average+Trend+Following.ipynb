```python
import pandas_datareader.data as wb, pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn
from datetime import datetime
%matplotlib inline
```

```python
spxl = wb.DataReader('SPXL', data_source = 'yahoo', start = datetime(2015,6,24))
```

```python
data = pd.DataFrame({'spxl':spxl['Adj Close']})
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)   # rows, columns, # location


data.plot(ax = ax1, figsize = (22,10))

data.rolling(window = 10).mean().plot(ax = ax1)
data.rolling(window = 30).mean().plot(ax = ax1)

plt.legend(['SPXL','10 day SMA', '30 day SMA']);
```

```python
# Let's See the SPXL(SPY's 3 times leveage ETF)'s sma_10_day - sma_30_day
sma10 = data.rolling(window = 10).mean()
sma30 = data.rolling(window = 30).mean()

sma_spread = sma10-sma30
sma_spread.plot(figsize = (22,8)).legend(["sma10-sma30 Spread"])
```
