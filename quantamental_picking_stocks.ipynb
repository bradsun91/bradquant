- 1. 市盈(TTM)：Price Earnings Ratio；股票价格／每股收益；越小越好

- 2. 市净率：Price-to-Book Ratio:，每股股价／每股净资产；越低，投资价值越高

- 3. 市现率：股票价格/每股现金流量；越低，说明每股现金增加额越多，经营压力越小

- 4. 市销率：Price-to-Sales Ratio，每股股价／每股销售额；越低，说明该公司股票的投资价值越大

- 5. 股息率%：Dividend Yield Ratio，股息／股票价格；是一年的总派息额与当时市价的比例。该指标是投资收益率的简化形式；越大越好

- 6. 每股收益：越大越好

- 7. 每股净资：越大越好

- 8. 每股现金流：越大越好

- 9. 净益率（净资产收益率）%：ROE；该指标体现了自有资本活得净收益的能力；指标越高，说明投资带来的收益越高。

- 10. 营业利润率%：Operating Profit Ratio；营业利润／全部业务收入*100%；越大越好

- 11. 净利润率%：净利润/营业收入×100%；越大越好

```python
def ensure_float(x):
    if isinstance(x,np.float):
        return x
    else :
        return np.NaN
```

# 0. 先按照流通市值，得出每个行业中的市值排名

```python
fin_name0 = "流通市值"
rank_col_name0 = fin_name0+"排名"
score_col_name0 = fin_name0+"排名分数"

fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: str(x))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.replace(" ",""))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.replace("亿","00000000"))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.replace(".",""))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.replace("--",""))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.strip("\n"))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.strip("\t"))
fin_data[fin_name0] = fin_data[fin_name0].apply(lambda x: x.strip("\""))

fin_data = fin_data.replace(r'', np.NaN)
for i, item in enumerate(fin_data[fin_name0]):
    try:
        fin_data[fin_name0][i] = float(fin_data[fin_name0][i])
    except:
        # to see if there's any value/row that cannot be converted to float
        print(fin_data[fin_name0][i], i)
        
        
fin_data[fin_name0] = fin_data[fin_name0].apply(ensure_float)
fin_data[rank_col_name0] = fin_data.groupby(['细分行业'])[fin_name0].rank(ascending = False)
fin_data[fin_data['细分行业']=="全国地产"][['代码','名称','细分行业',fin_name0,rank_col_name0]].sort_values(rank_col_name0)
```

# 技术指标 - 周线+日线

```python
def MACD(df, n_fast, n_slow, n_macd, price_col): # n_fast = 12, n_slow = 26
    """
    http://stockcharts.com/docs/doku.php?id=scans:indicators
    MACD, MACD Signal and MACD difference, rationale CHECKED, code CHECKED, updated
    # Conventional look-back window for calculating MACDsign is 9
    """
    EMAfast = df[price_col].ewm(span = n_fast, min_periods = n_fast - 1).mean()
    EMAslow = df[price_col].ewm(span = n_slow, min_periods = n_slow - 1).mean()
    MACD = pd.Series(EMAfast - EMAslow, name = 'MACD_' + str(n_fast) + '_' + str(n_slow))
    MACDsign = MACD.ewm(span = n_macd, min_periods = n_macd-1).mean().rename('MACDsign_' + str(n_fast) + '_' + str(n_slow))
    MACDdiff = pd.Series(MACD - MACDsign, name = 'MACDdiff_' + str(n_fast) + '_' + str(n_slow))
    df['MACD_Diff'] = MACD
    df['MACD_Diff_EMA'] = MACDsign
    df['MACD'] = MACDdiff
    df['SIGNAL_STATUS'] = df['MACD'].apply(lambda x: "多头状态" if x>0 else ("空头状态" if x<0 else "无信号状态"))
    return df

indicator_df = MACD(df, 12, 26, 9, 'close')
indicator_df.index = pd.to_datetime(indicator_df['trade_date'])
```

```python
# 查看周线：

period_type = "W"
df_copy = df.copy()
weekly_df = df_copy.resample(period_type).last()
weekly_indicator_df = MACD(weekly_df, 12, 26, 9, 'close')
```

```python
last_daily_signal = indicator_df['SIGNAL_STATUS'].values[-1]
last_weekly_signal = weekly_indicator_df['SIGNAL_STATUS'].values[-1]
last_price_vol_date = indicator_df['trade_date'].values[-1]
```
