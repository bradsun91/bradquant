```python
import docx
import xlrd
import random
import re
from word import *
```

```python
#  ---------------------------
# __author__ = 'stormsha'
# __date__ = '2019/2/2 9:34'
# ---------------------------
class MyDocument(object):

    def __init__(self):
        self.data = {}
        self.file_name = "./automation_files/"
        self.paragraphs = self.read_template()
        self.module_data = self.read_module()

    # 读取模板：
    def read_template(self):
        template = docx.Document(self.file_name + "2_18_2019_v4_自动化word模板读取.docx")
        print (len(template.paragraphs))
#         print("段落数:" + str(len(template.paragraphs)))      # 段落数为13，每个回车隔离一段
        for i in range(len(template.paragraphs)):
            print("第" + str(i) + "段的内容是：" + template.paragraphs[i].text)
        return template.paragraphs

    # 读取组合好的文章
    def read_module(self):
        workbook = xlrd.open_workbook(self.file_name + "module.xlsx")
        sheet_names = workbook.sheet_names()
        data = sheet_names[0]
        sheet1 = workbook.sheet_by_name(data)
        row_num = sheet1.get_rows()
        module_list = []
        module = {}
        for index, row in enumerate(row_num):
            if index > 0:
                module["id"] = row[0].value
                module["keyword1"] = row[1].value
                module["keyword2"] = row[2].value
                module['keyword3'] = row[3].value
                module['keyword4'] = row[4].value
                module["module1"] = row[5].value
                module["module2"] = row[6].value
                module["module3"] = row[7].value
                module["module4"] = row[8].value
                module_list.append(module)
            module = {}
        return module_list

    # 读取变量
    def read_var(self, index_name):
        workbook = xlrd.open_workbook(self.file_name + "var.xlsx")
        sheet_names = workbook.sheet_names()
        data = sheet_names[0]
        sheet1 = workbook.sheet_by_name(data)
        headers = sheet1.row_values(0)
        num = headers.index(index_name)
        col_values = sheet1.col_values(num)[1:]
        cols = []
        for index, x in enumerate(col_values):
            if x == "NA":
                pass
            else:
                cols.append(x)
        var = random.choices(cols[1:])
        print (var[0])
        try:
            return str(var[0])
        except (UnboundLocalError, Exception):
            pass

    # 替换变量
    def group(self, module):
        print(module)
        for i in range(len(self.paragraphs)):
            print ('=========================================================')
            print (i, self.paragraphs[i].text)
            self.data[i] = self.replace_content(self.paragraphs[i].text, module)
            self.data[i] = self.replace_content(self.data[i], module)
        # 生成文章
        word(self.data)
        self.data = {}

    # 替换变量var或者var_
    def replace_content(self, text, module):
        var_all = re.findall(r'#(.*?)#', text)
        for var in var_all:
            number = re.findall(r'(\d+)', var)[0]
            category = re.sub(number, '', var)
            if category == "var" or category == "var_" or category == "title_":
#                 print("变量：", self.read_var(int(number)))
#                 print ("***************************************************")
#                 print ("number", number)
                print(var)
                text = text.replace('#' + var + '#', self.read_var(var))
            else:
                print("VAR+MODULE", var, module[var])
                text = text.replace('#' + var + '#', module[var])
        print ("替换之后的text", text)
        print("================================================================")
        return text

    # 执行组合功能
    def deal(self):
        for module in self.module_data:
            self.group(module)


# 主程序
if __name__ == '__main__':
    document = MyDocument()
    document.read_template()
    document.deal()

    
#     str_ = "#var_1#最近股市波动不断，#var_2#无论是#var_3#相同的还是不同市场，交易的#var4#分析手段上都是有#var5#共性的，这里#var6#小编我就给#var7#各位介绍一下有经验的#keyword1#交易员#var8#应对#keyword2#股票市场的#var9#秘密武器 - #keyword3#MACD。"
#     module = {}
#     text = document.replace_content(str_, module)
#     print(text)
```
