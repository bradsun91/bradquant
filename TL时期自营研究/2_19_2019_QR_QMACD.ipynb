```python
eos = pd.read_csv(loc+"eos15min.csv", engine="python", header = None)
eth = pd.read_csv(loc+"eth15min.csv", engine="python", header = None)

eos.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
eth.columns = ['date', 'open', 'high', 'low', 'close', 'volume']

eos['date'] = pd.to_datetime(eos['date'])
eth['date'] = pd.to_datetime(eth['date'])

eos.index = eos['date']
eth.index = eth['date']

eos = eos.iloc[1:]
eth = eth.iloc[1:]
```

```python
ltc = pd.read_csv(loc+"ltc15min.csv", engine="python", header = None)
ltc.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
ltc['date'] = pd.to_datetime(ltc['date'])
ltc.index = ltc['date']
ltc = ltc.iloc[1:]
```

```python
def process_daily(file):
    asset = pd.read_csv(loc+file, engine="python", header = None)
    asset.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
    asset['date'] = pd.to_datetime(asset['date'])
    asset.index = asset['date']
    asset = asset.iloc[1:]
    daily_asset = asset.resample('D').last()
    return daily_asset
```

```python
daily_ltc = process_daily("ltc15min.csv")
```
