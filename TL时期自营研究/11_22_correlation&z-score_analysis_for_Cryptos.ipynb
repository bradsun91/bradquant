```python
def reset_price(location, till_date, file_suffix, added_note):

    symlist = ['ada', 'bch', 'eth', 'eos', 'trx', 'xrp', 'ltc']

    for s in symlist:
        if s == 'ada':
            c = 10000000
        elif s == 'bch':
            c = 1000
        elif s == 'eos':
            c = 100000
        elif s == 'eth' or s == 'ltc':
            c = 10000
        elif s == 'trx':
            c = 100000000
        elif s == 'xrp':
            c = 10000000

#         b = pd.read_csv(location + till_date + s + file_suffix,header=None)
        b = pd.read_csv(location + s + file_suffix + ".csv")
        b.iloc[:, [1, 2, 3, 4]] = b.iloc[:, [1, 2, 3, 4]].astype(float)
        b.iloc[:, [1, 2, 3, 4]] = (b.iloc[:, [1, 2, 3, 4]]) * c
        b.to_csv(location + 'res_' + till_date + s + added_note + str(c) + '.csv', 
                 sep=',', header=False, index=False, float_format='%.2f')

till_date = "11_21_"
file_suffix = 'u18z18_1d'
added_note = 'u18z18乘'

reset_price(location, till_date, file_suffix, added_note)
```

# Calculate rolling z-scores of close price pct_change()

## ADA and ETH

```python
ada_eth['diff_price_pct'] = ada_eth['ada_pct_chg']-ada_eth['eth_pct_chg']
```

```python
def zscore(x, window):
    r = x.rolling(window=window)
    m = r.mean().shift(1)
    s = r.std(ddof=0).shift(1)
    z = (x-m)/s
    return z

def create_ada_profits(signal_nextd, ada_close_pct):
    # we want to make money by going long on ada when the signal == -1(signal < 0)
    if signal_nextd == -1 and ada_close_pct > 0:
        # made money
        return ada_close_pct
    elif signal_nextd == -1 and ada_close_pct < 0:
        # lost money
        return -abs(ada_close_pct)
    elif signal_nextd == 1 and ada_close_pct > 0:
        # lost money
        return -ada_close_pct
    elif signal_nextd == 1 and ada_close_pct < 0:
        # made money
        return abs(ada_close_pct)
    
def create_eth_profits(signal_nextd, eth_close_pct):
    # we want to make money by going short on eth when the signal == -1(signal < 0)
    if signal_nextd == -1 and eth_close_pct > 0:
        # lost moeny 
        return -eth_close_pct
    elif signal_nextd == -1 and eth_close_pct < 0:
        # made money
        return abs(eth_close_pct)
    elif signal_nextd == 1 and eth_close_pct > 0:
        # made money
        return eth_close_pct
    elif signal_nextd == 1 and eth_close_pct < 0:
        # lost money
        return -abs(eth_close_pct)
```

```python
# paras setup:

rolling_days = 33
z_top = 1
z_bottom = -1

ada_eth = ada_eth[['ada', 'eth', 'diff_pct', 'ada_pct_chg', 'eth_pct_chg',
       'diff_price_pct']]

ada_eth['z_score_{}d'.format(rolling_days)] = zscore(ada_eth.diff_price_pct, rolling_days)

# If z-score is too low, long ada and short eth, for now the threshold can be set as 1 and -1: 
ada_eth['signal'] = ada_eth['z_score_{}d'.format(rolling_days)].apply(lambda x: z_bottom if x <=z_bottom else(z_top if x >=z_top else 0))
ada_eth['signal_for_next_d_calc'] = ada_eth['signal'].shift(1)

ada_eth['ada_profits'] = np.vectorize(create_ada_profits)(ada_eth['signal_for_next_d_calc'], ada_eth['ada_pct_chg'])

ada_eth['eth_profits'] = np.vectorize(create_eth_profits)(ada_eth['signal_for_next_d_calc'], ada_eth['eth_pct_chg'])

ada_eth['ada_sum_equity'] = ada_eth['ada_profits'].cumsum()
ada_eth['eth_sum_equity'] = ada_eth['eth_profits'].cumsum()
ada_eth['total_sum_equity'] = ada_eth['ada_sum_equity'] + ada_eth['eth_sum_equity']

ada_eth.dropna()['total_sum_equity'].plot()
```
