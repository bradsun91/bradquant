```python
# =========================================Strategy 1：macd_updown_ma_slope_bias_all_df=======================================

while True:  
# ==============================================================================
# ==============================================================================
    print("====================Start Excuting Strategy 1!!!====================")

    # Fill US stocks:

    # 找到账户里的ticker：
    position_num = len(ib.positions())
    positions = []
    for i in range(0, position_num):
        pos_ticker = ib.positions()[i].contract.symbol
        positions.append(pos_ticker)
        print(pos_ticker)

    us_db_path = "C:/Users/Administrator/CE_github_2/data_pipeline/Data/"
    data_list_selected = LOAD_DATA.read_data_from_yf_60m(ticker_list)

    # Generate signals
    last_ts_signals, all_signal_df = GENERATE_SIGNALS.generate_signals_macd_updown_ma_slope_bias( 
                                                                           data_list_selected,
    #                                                                            CN_fund_data,
                                                                          "Macd_Updown_Ma_Slope_Bias", 
                                                                          "Datetime", 
                                                                          "Adj Close", 
                                                                          "Ticker", 
                                                                          "1h")
    last_ts_signals['account_value'] = ib.accountValues()[124][2]
    # print(last_ts_signals)
    # futures
    for ticker in ticker_list:
        IB_ticker = YH_IB_Ticker_Map_US_Index_Futures[ticker]
        ticker_pos_direction = YH_IB_Ticker_Map_US_Index_Futures_Directions[IB_ticker]
        print("检测{}信号...".format(ticker))

        # 若没有仓位：
        if ticker_pos_direction==0:
            # 进多仓信号，看情况启动：
            if str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="买入" or \
                str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="多头持有":
    #             print("Signal for: {}".format(ticker),":买入")

                contract = Future(IB_ticker, YH_IB_Ticker_Map_US_Index_Futures_Exp_Dates[IB_ticker], YH_IB_Ticker_Map_US_Index_Futures_Exchanges[IB_ticker])
                ib.qualifyContracts(contract)
    ##================================================================
#     #                 step 2 定义合约单信息
#                 order = MarketOrder('BUY', YH_IB_Ticker_Map_US_Index_Futures_Allocation[IB_ticker])

#     # #                 # step 3 定义交易功能
#                 trade = ib.placeOrder(contract, order)
    ##================================================================
    #                 print(trade)

                ib.sleep(1)
                print("Bought to Open Pos")
                send_email(last_ts_signals)      

            # 平仓信号，不启动：
            elif str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="卖出平仓" or \
            str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="无仓位":
                
                print("No positions to sell close")

            else:
                print("No positions; No Actions")


        # 若有多头仓位：
        elif ticker_pos_direction>0:
#             # 进多仓信号，不启动
            if str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="买入" or \
                str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="多头持有":   
                print("Already has the same position.")


            # 卖出/平仓信号，看情况启动： 
            if str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="卖出平仓":
                contract = Future(IB_ticker, YH_IB_Ticker_Map_US_Index_Futures_Exp_Dates[IB_ticker], YH_IB_Ticker_Map_US_Index_Futures_Exchanges[IB_ticker])
                ib.qualifyContracts(contract)
    ##================================================================
#                 # step 2 定义合约单信息
#                 order = MarketOrder('SELL', YH_IB_Ticker_Map_US_Index_Futures_Allocation[IB_ticker])

#     #                 # step 3 定义交易功能
#                 trade = ib.placeOrder(contract, order)
    ##================================================================
                # print(trade)

                ib.sleep(1)
                print("Sold to Close Pos.")       
                send_email(last_ts_signals)   

            else:
                print("No actions")

        print("--------------")
        
# ==============================================================================
# ==============================================================================
    time.sleep(60)
```

```python
while True:  
# ==============================================================================
# ==============================================================================

    print("====================Start Excuting Strategy 2!!! Version of 20210622====================")

    # Fill US stocks:

    # 找到账户里的ticker：
    position_num = len(ib.positions())
    positions = []
    for i in range(0, position_num):
        pos_ticker = ib.positions()[i].contract.symbol
        positions.append(pos_ticker)
        print(pos_ticker)

    us_db_path = "C:/Users/Administrator/CE_github_2/data_pipeline/Data/"
    data_list_selected = LOAD_DATA.read_data_from_yf_60m(ticker_list)



    # Generate signals
    last_ts_signals, all_signal_df = GENERATE_SIGNALS.generate_signals_cross_down_5ma_short_1h( 
                                                                           data_list_selected,
                                                                          "cross_down_5ma_short_1h", 
                                                                          "Datetime", 
                                                                          "Adj Close", 
                                                                          "Ticker", 
                                                                          "1h")
    last_ts_signals['account_value'] = ib.accountValues()[124][2]
    # print(last_ts_signals)
    # futures
    for ticker in ticker_list:
        IB_ticker = YH_IB_Ticker_Map_US_Index_Futures[ticker]
        ticker_pos_direction = YH_IB_Ticker_Map_US_Index_Futures_Directions[IB_ticker]
        print("检测{}信号...".format(ticker))

        # ==========若没有仓位 or 有多头仓位：==========
        if ticker_pos_direction==0 or ticker_pos_direction>0:
            # 做空进仓信号：
            if str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="做空" or \
                str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="空头持有":
    #             print("Signal for: {}".format(ticker),":买入")

                contract = Future(IB_ticker, '20210917', YH_IB_Ticker_Map_US_Index_Futures_Exchanges[IB_ticker])
                ib.qualifyContracts(contract)
    ##================================================================
#         #                 step 2 定义合约单信息
#                     order = MarketOrder('SELL', YH_IB_Ticker_Map_US_Index_Futures_Allocation[IB_ticker])

#         # #                 # step 3 定义交易功能
#                     trade = ib.placeOrder(contract, order)
    ##================================================================
    #                 print(trade)

                ib.sleep(1)
                print("Sold Short Pos")
                send_email(last_ts_signals)      

            # 买入平仓信号：
            elif str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="买入平仓" or \
            str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="-" or \
            str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="无仓位":
                # 若没有仓位：
                print("No positions to buy close")

            else:
                print("No Actions")


        # ==========若有空头仓位：==========
        elif ticker_pos_direction <0:


            # 做空进仓信号：
            if str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="做空" or \
                str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="空头持有":
                print("Already has the same short position.")


            # 买入/平仓信号： 
            elif (str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="买入平仓" or \
                str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="-" or \
                str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="无仓位"):

                contract = Future(IB_ticker, '20210917', YH_IB_Ticker_Map_US_Index_Futures_Exchanges[IB_ticker])
                ib.qualifyContracts(contract)
    ##================================================================
    #                 # step 2 定义合约单信息
    #                 order = MarketOrder('BUY', YH_IB_Ticker_Map_US_Index_Futures_Allocation[IB_ticker])

    #     #                 # step 3 定义交易功能
    #                 trade = ib.placeOrder(contract, order)
    ##================================================================
                # print(trade)

                ib.sleep(1)
                print("Bought to Close Pos.")       
                send_email(last_ts_signals)   

            else:
                print("No actions")

        print("--------------")
        
# ==============================================================================
# ==============================================================================
    time.sleep(3600)
```
