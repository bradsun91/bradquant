```python
while True:
    print("Start Excuting!!!")

    # Fill US stocks:

    # 找到账户里的ticker：
    position_num = len(ib.positions())
    positions = []
    for i in range(0, position_num):
        pos_ticker = ib.positions()[i].contract.symbol
        positions.append(pos_ticker)
        print(pos_ticker)

    us_db_path = "C:/Users/Administrator/CE_github_2/data_pipeline/Data/"
    data_list_selected = LOAD_DATA.read_data_from_yf_60m(ticker_list)

    # Generate signals
    last_ts_signals, all_signal_df = GENERATE_SIGNALS.generate_signals_macd_updown_ma_slope_bias( 
                                                                           data_list_selected,
    #                                                                            CN_fund_data,
                                                                          "Macd_Updown_Ma_Slope_Bias", 
                                                                          "Datetime", 
                                                                          "Adj Close", 
                                                                          "Ticker", 
                                                                          "1h")

    print(last_ts_signals)
    # futures
    for ticker in ticker_list:
        print("检测{}信号...".format(ticker))
        if str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="买入" or \
            str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="多头持有":
#             print("Signal for: {}".format(ticker),":买入")
            IB_ticker = YH_IB_Ticker_Map_US_Index_Futures[ticker]

            # 若没有仓位：
            if IB_ticker not in positions:

                contract = Future(IB_ticker, '20210618', YH_IB_Ticker_Map_US_Index_Futures_Exchanges[IB_ticker])
                ib.qualifyContracts(contract)
                # step 2 定义合约单信息
                order = MarketOrder('BUY', 2)

                # step 3 定义交易功能
                trade = ib.placeOrder(contract, order)

                # print(trade)

                ib.sleep(1)
                print("Bought to Open Pos")
                send_email(last_ts_signals)

            # 若已经有仓位：    
            else:
                print("Already has the same position.")

        elif str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="卖出平仓" or \
            str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="-" or \
            str(last_ts_signals[last_ts_signals['Ticker'] == ticker]['Signal'].values[0])=="无仓位":
#             print("Signal for: {}".format(ticker),":卖出平仓")
            IB_ticker = YH_IB_Ticker_Map_US_Index_Futures[ticker]


            # 若有仓位：
            if IB_ticker in positions:
                contract = Future(IB_ticker, '20210618', YH_IB_Ticker_Map_US_Index_Futures_Exchanges[IB_ticker])
                ib.qualifyContracts(contract)
                # step 2 定义合约单信息
                order = MarketOrder('SELL', 2)

                # step 3 定义交易功能
                trade = ib.placeOrder(contract, order)

                # print(trade)

                ib.sleep(1)
                print("Sold to Close Pos.")       
                send_email(last_ts_signals)

            # 若没有仓位：
            else:
                print("No positions to sell close")
        else:
            print("No actions")
            pass
        print("--------------")
    time.sleep(3600)
```
