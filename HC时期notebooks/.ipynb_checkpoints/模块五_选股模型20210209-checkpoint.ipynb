# 过去252交易日收益率排名

```python
def calculate_1yr_return(df, price_col, date_col, ticker_col):
    df = df.tail(252)
    rtrn = df[price_col].values[-1]/df[price_col].values[0] - 1
    ticker = df[ticker_col].values[-1]
    last_date = df[date_col].values[-1]
    first_date = df[date_col].values[0]
    return rtrn, ticker, last_date, first_date
```

```python
all_csvs = "C:/Users/Administrator/CE_github_2/data_pipeline/Data/*.csv"
stock_list = []
date_list = []
return_list = []
len_ = 1
date_col = "Date"
price_col = "Adj Close"
model_freq = "D"
return_df = pd.DataFrame()
model_name = "Last_252_days_Returns"
total_number = len(glob.glob(all_csvs)[:])
ticker_col = "Ticker"


for fname in glob.glob(all_csvs)[:]:
    print ("Calculating 252-Day Returns for No.{} : {}".format(len_, fname))
    data = pd.read_csv(fname)
    try:
        data_252_rtrn, ticker, last_date, first_date = calculate_1yr_return(data, price_col, date_col, ticker_col)
        stock_list.append(ticker)
        date_list.append(last_date)
        return_list.append(data_252_rtrn)
    except Exception as e:
        print(e)
    len_+= 1
return_df['ticker'] = stock_list
return_df['first_date'] = first_date
return_df['last_date'] = date_list
return_df['returns'] = return_list
return_df['model_name'] = model_name
return_df['model_freq'] = model_freq
return_df_daily = return_df.copy()
print("All Done!")
```

```python
return_df_daily.sort_values("returns", ascending=False, inplace = True)
```

```python
return_df_daily['Return_Rank'] = return_df_daily['returns'].rank(ascending=False)
```

# 过去252天夏普率排名

```python
def calculate_1yr_sharpe(df, price_col, date_col, ticker_col):
    """
    Documentation:
    ---------
    Strategy_ret: Series | Numpy Array
    Freq_multiplier explanation:
    Daily_Data_Freq_multiplier: 252, 
    Hourly_Data_Freq_multiplier: 252*6.5,  
    Minutely_Data_Freq_multipler: 252*6.5*60 
    Second_Data_Freq_multipler: 252*6.5*60*60
    """
    df = df.tail(252)
    ticker = df[ticker_col].values[-1]
    last_date = df[date_col].values[-1]
    first_date = df[date_col].values[0]
    strategy_ret = df[price_col].pct_change().dropna()
    daily_mean_ret = np.mean(strategy_ret)
    daily_stdev = np.std(strategy_ret, ddof=1)
    sharpe = daily_mean_ret/daily_stdev*np.sqrt(252)
    return sharpe, ticker, last_date, first_date
```

```python
all_csvs = "C:/Users/Administrator/CE_github_2/data_pipeline/Data/*.csv"
stock_list = []
date_list = []
sharpe_list = []
len_ = 1
date_col = "Date"
price_col = "Adj Close"
model_freq = "D"
sharpe_df = pd.DataFrame()
model_name = "Last_252_days_Sharpe"
total_number = len(glob.glob(all_csvs)[:])
ticker_col = "Ticker"


for fname in glob.glob(all_csvs)[:]:
    print ("Calculating 252-Day Sharpe for No.{} : {}".format(len_, fname))
    data = pd.read_csv(fname)
#     print(data.head(2))
    try:
        data_252_sharpe, ticker, last_date, first_date = calculate_1yr_sharpe(data, price_col, date_col, ticker_col)
        stock_list.append(ticker)
        date_list.append(last_date)
        sharpe_list.append(data_252_sharpe)
    except Exception as e:
        print(e)
    len_+= 1

    
sharpe_df['ticker'] = stock_list
sharpe_df['first_date'] = first_date
sharpe_df['last_date'] = date_list
sharpe_df['sharpe'] = sharpe_list
sharpe_df['model_name'] = model_name
sharpe_df['model_freq'] = model_freq
sharpe_df_daily = sharpe_df.copy()
print("All Done!")
```

```python
sharpe_df_daily.sort_values("sharpe", ascending=False, inplace = True)
```
