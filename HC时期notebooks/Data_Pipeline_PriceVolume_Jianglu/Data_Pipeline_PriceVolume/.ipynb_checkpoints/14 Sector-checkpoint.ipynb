```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
import seaborn as sns

general_path = "C:/Users/Administrator/CE_github_2/data_pipeline/Data/"
sectors_list = ["XLE","XLV","XLI","XLF","XLY","XLP","XLB","XLK","XLU","XLRE","XLC","SHY","IEF","TLT"]

def read_csv (sector):
    return  pd.read_csv(general_path + sector +".csv")
```

## 计算daily percentage change

```python
def daily_percentage_change (data):
    return data['Adj Close']/data['Adj Close'].shift(1)-1
```

## 计算sector与大盘变化的差

```python
12065.09/13241.48-1
```

```python
SPY = read_csv("SPY")
spy_daily =  daily_percentage_change(SPY)
def change_diff(sector):
    return daily_percentage_change(sector) - spy_daily
```

## 整合数据
* 14 sectors 的数据整合
* 与大盘数据差的整合

```python
sectors = pd.DataFrame() # 14 sectors
sectors_dif = pd.DataFrame()# 数据差

sectors["SPY"] = SPY["Adj Close"]

for sector in sectors_list:
    data = read_csv(sector)
    sectors[sector] = data["Adj Close"]
    sectors_dif[sector] = change_diff(data)

sectors.index = SPY["Date"]   
sectors_dif.index = SPY["Date"]
```

```python
sectors = sectors.fillna(0)
sectors
```

# Calculate Z-scores

```python
def get_z_score(series):
    rol_mean = series.rolling(window = 10).mean()
    rol_std = series.rolling(window = 10).std()
    scores = [(series-rol_mean)/rol_std]
    return scores
```

```python
get_z_score(sectors_dif)[0][['XLF','XLE']].tail(100).plot()
```

```python
sectors_dif.tail(100).plot(figsize = (18,6))
```
